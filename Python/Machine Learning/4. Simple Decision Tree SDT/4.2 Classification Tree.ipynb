```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# classification Tree
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
# Para graficar el Decision Tree
from IPython.display import Image
import pydotplus
import os
```

```python
df = pd.read_csv('./Archivos csv/Movie_classification.csv', header = 0)
```

```python
df.head()
```

```python
df.info()
```

# Valores Faltantes

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(), inplace=True)
```

```python
df.info()
```

# Creacion variable Dummies

```python
df =pd.get_dummies(df, columns=['3D_available','Genre'], drop_first = True)
```

```python
df.info()
```

# X-y Split

```python
X = df.loc[:, df.columns != 'Start_Tech_Oscar']
type(X)
```

```python
X.shape
```

```python
y = df['Start_Tech_Oscar']
type(y)
```

```python
y.head()
```

```python
y.shape
```

# Train-Test-Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Entrenando el Arbol de Clasificacion

```python
clftree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clftree.fit(X_train, y_train)
```

# Prediccion de valores usando el modelo

```python
y_train_pred = clftree.predict(X_train)
y_test_pred  = clftree.predict(X_test)
```

# Redimiento del modelo

```python
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

# Graficando el Arbol de Clasificacion

```python
dot_data = tree.export_graphviz(clftree, out_file =None, feature_names=X_train.columns, filled=True)
```

```python
os.environ["PATH"] += os.pathsep + 'C:\Program Files\Graphviz/bin/'
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

# Controlando el Crecimiento del Arbol

```python
clftree1 = tree.DecisionTreeClassifier(min_samples_split=20, max_depth=4)
clftree1.fit(X_train, y_train)
dot_data = tree.export_graphviz(clftree1, out_file =None, feature_names= X_train.columns, filled =True)
graph1   = pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())  
```

```python
accuracy_score(y_test, clftree1.predict(X_test))
```
