```python
np.percentile(df.rainfall, [1])[0]
```

```python
lv = np.percentile(df.rainfall, [1])[0]
```

```python
df[(df.rainfall < lv)]
```

# Train-test split

```python
X_train, X_test, ytrain, ytest = train_test_split(X_multi, y_multi, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape, X_test.shape, ytrain.shape, ytest.shape)
```

```python
lm_a = LinearRegression()
```

```python
lm_a.fit(X_train, ytrain)
```

```python
y_test_a = lm_a.predict(X_test)
```

```python
y_train_a = lm_a.predict(X_train)
```

```python
r2_score(ytest, y_test_a)
```

```python
r2_score(ytrain, y_train_a)
```

# Regresion Ridge y Lasso

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
lm_r = Ridge(alpha = 0.5)
lm_r.fit(X_train_s, ytrain)
```

```python
r2_score(ytest, lm_r.predict(X_test_s))
```

```python
param_range = np.logspace(-2,8,100)
```

```python
param_range
```

```python
train_scores, test_scores = validation_curve(Ridge(), X_train_s, ytrain, param_name="alpha", param_range=param_range, scoring='r2')
```
