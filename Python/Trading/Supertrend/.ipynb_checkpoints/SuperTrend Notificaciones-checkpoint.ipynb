```python
import MetaTrader5 as mt5
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import matplotlib.dates as mpl_dates
import matplotlib.pyplot as plt

from datetime import datetime 
import schedule
import time

import pywhatkit
```

# Importando Datos MT5

```python
def Simbolo(divisa, tf,em,ter):  
           
    temp = {'M1':mt5.TIMEFRAME_M1,'M5':mt5.TIMEFRAME_M5,
            'M15':mt5.TIMEFRAME_M15,'M30':mt5.TIMEFRAME_M30,
            'H1':mt5.TIMEFRAME_H1,'H4':mt5.TIMEFRAME_H4
           }
    for clave, cont in temp.items():
        if tf == clave:
            rates = mt5.copy_rates_from_pos(divisa, cont, em, ter)

    # Transform Tuple into a DataFrame
    df_rates = pd.DataFrame(rates).dropna()
    

    # Convert number format of the date into date format
    df_rates["time"] = pd.to_datetime(df_rates["time"], unit="s")
    df_rates = df_rates.drop(['tick_volume','spread','real_volume'],axis=1)
    
    df_rates.time = pd.to_datetime(df_rates['time'])
    df_rates.index = df_rates.time
    df_rates.index.name = 'date'
    df_rates['time'] = df_rates['time'].apply(mpl_dates.date2num)

    return df_rates

symbol    = 'EURUSD'
timeframe = 'M1'
ini = 0
ter = 1000
df = Simbolo(symbol,timeframe,ini,ter)
df.head()
```

# SuperTrend (ATR)

```python
def SuperTrend(df):
    #ATR
    high_low = np.abs(df['high'] - df['low'])
    high_pc = np.abs(df['high'] - df['close'].shift(1))
    low_pc = np.abs(df['low'] - df['close'].shift(1))

    ranges = pd.concat([high_low, high_pc, low_pc],axis=1)
    true_range = np.max(ranges, axis=1)
    df['atr'] = true_range.rolling(7).mean() #14
    
    multiplier = 3
    df['upperband'] = ((df.high + df.low) / 2) + (multiplier * df.atr)
    df['lowerband'] = ((df.high + df.low) / 2) - (multiplier * df.atr)
    df['in_uptrend'] = True
    
    for current in range(1, len(df.index)):
        previous = current - 1

        if df['close'][current] > df['upperband'][previous]:
            df['in_uptrend'] = True
        elif df['close'][current] < df['lowerband'][previous]:
            df['in_uptrend'] = False
        else:
            df['in_uptrend'][current] = df['in_uptrend'][previous]

            if df['in_uptrend'][previous] and df['lowerband'][current] < df['lowerband'][previous]:
                df['lowerband'][current] = df['lowerband'][previous]

            if not df['in_uptrend'][current] and df['upperband'][current] > df['upperband'][previous]:
                df['upperband'][current] = df['upperband'][previous]
    
    return df

st = SuperTrend(df)
st.tail(10)
```

```python
def mensaje(symbol,texto):
    # Usamos Un try-except
    try:
        # Enviamos el mensaje
        if texto == 'Compra':
            day = datetime.now().isoformat()
            pywhatkit.sendwhatmsg("+573002479014",
                                f"COMPRA en {symbol}",
                                int(day[11:13]),int(day[14:16])+1)
            return print("Mensaje Enviado")
        else:
            day = datetime.now().isoformat()
            pywhatkit.sendwhatmsg("+573002479014",
                                f"VENTA en {symbol}",
                                int(day[11:13]),int(day[14:16])+1)
            return print("Mensaje Enviado")

    except:
        return print("Ocurrio Un Error")

```

```python
posicion = False 

def señales_compra_o_venta(df, symbol):
    global posicion
      
    print('buscando señales..........','\n\n')
    
    ultima_fila    = len(df.index) - 1
    penultima_fila = ultima_fila - 1
    
    if not df['in_uptrend'][penultima_fila] and df['in_uptrend'][ultima_fila]:
        print('Cambio de tendencia, COMPRAR','\n')
        if not posicion:
            mensaje(symbol,'Compra')
            posicion = True
        else:
            print('Ya se encuentra en posicion de COMPRA')
           
    if df['in_uptrend'][penultima_fila] and not df['in_uptrend'][ultima_fila]:
        print('Cambio de tendencia, VENTA','\n')
        if posicion:
            mensaje(symbol,'Venta')
            posicion = False
        else:
            print('Ya se encuentra en posicion de VENTA')
    
```

```python
def run_bot():
#     print(f'Buscando barras nuevas desde: {datetime.now().isoformat()}','\n')
    data   = Simbolo(symbol,timeframe,0,50)
    print(f'Buscando barras nuevas desde: {data.index[-1]}','\n')
    STdata = SuperTrend(data)
    señales_compra_o_venta(STdata, symbol)

# temp = {'M1':1,'M5':5,'M15':15,'M30':30,'H1':60,'H4':240}

# for clave, contenidos in temp.items():
#     if timeframe == clave: 
#         schedule.every(contenidos).minutes.do(run_bot)
schedule.every(1).minute.do(run_bot)

while True:
    schedule.run_pending()
    
    time.sleep(1)
```
