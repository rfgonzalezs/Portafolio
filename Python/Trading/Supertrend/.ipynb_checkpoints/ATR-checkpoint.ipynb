```python
import ta
import yfinance as yf
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import matplotlib.dates as mpl_dates
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = [20, 10]
```

# Importando Datos YFinance

```python
def preprocessing(symbol, start,end):  
    #Import the data
    df = yf.download(symbol,start,end).dropna()

    #Rename
    df.columns = ["open", "high", "low", "close", "adj close", "volume"]
    df.index.name = "date"
    df['time'] = df.index

    # Remove adj close
    del df["adj close"]
    del df["volume"]

    return df


symbol='EURUSD=X'
df = preprocessing(symbol, start="2020-01-01", end="2023-04-25")
df['time'] = df['time'].apply(mpl_dates.date2num)
df.head(10)
```

```python
high_low = df['high']-df['low']
high_close = np.abs(df['high']-df['close'].shift())
low_close = np.abs(df['low']-df['close'].shift())

ranges = pd.concat([high_close, high_low, high_close],axis=1)
true_range = np.max(ranges, axis=1)
atr = true_range.rolling(15).mean()
```

```python
fig, (ax1,ax2) = plt.subplots(2)
fig.suptitle(symbol[0:len(symbol)-2])
ax1.get_xaxis().set_visible(False)
df['close'].loc['2023'].plot(ax=ax1)
atr.loc['2023'].plot(ax=ax2, color='purple', ylabel='ATR')
plt.show()
```
